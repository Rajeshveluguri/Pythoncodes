```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img2 = np.zeros(shape = (500,500), dtype = np.int32)
plt.imshow(img2, cmap = 'gray')
```

Place image on the black box using PUT method

```python
cv2.putText(img2, "5", (100,400), cv2.FONT_HERSHEY_PLAIN, fontScale = 30, color = (255,255,255), thickness = 30)
plt.imshow(img2, cmap = 'gray')
```

```python
cv2.putText(img2, "RAJESH", (100,450), cv2.FONT_HERSHEY_PLAIN, fontScale = 5, color = (255,255,255), thickness = 5)
plt.imshow(img2, cmap = 'gray')
```

Same diagram in much stylish way
