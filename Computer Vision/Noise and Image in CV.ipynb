```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
noise = np.random.randint(low = 0, high = 2, size=(500,500))
```

```python
noise.shape
```

```python
plt.imshow(noise)
```

```python
plt.imshow(noise, cmap = 'gray')
```

```python
noise
```

```python
noise1 = noise*255
noise1
```

```python
plt.imshow(noise1, cmap = 'gray')
```

```python
img = np.zeros(shape = (500,500), dtype = np.int32)
plt.imshow(img, cmap = 'gray')
```

```python
cv2.circle(img, center = (250,250), radius = 50, color = (255,0,0), thickness = 5)
plt.imshow(img, cmap = 'gray')
```

### To induce noise into the image, we need to just add noise to image

```python
img1 = img + noise1
plt.imshow(img1, cmap = 'gray')
```

Noise only inside the image

```python
noise2 = np.random.randint(low = 0, high = 2, size = (500,500))
noise2
```

```python
noise2 = noise2 * -255
```

```python
noise2
```

```python
img2 = np.zeros(shape = (500,500), dtype = np.int32)
cv2.circle(img2, center = (250,250), radius = 50, color = (255,0,0), thickness = 5)
plt.imshow(img2, cmap = 'gray')
```

```python
img3 = img2 + noise2
img3
plt.imshow(img3, cmap = 'gray')
```

```python
img3[img3 == -255]=0
plt.imshow(img3, cmap = 'gray')
```

Above image is circle with noise.
