Polygons

```python
img6 = np.zeros(shape = (500,500,3), dtype = np.int32)
plt.imshow(img6)
```

```python
vertices = np.array([[100,300], [200,200], [400,300], [200,400]], np.int32)
```

```python
cv2.polylines(img6, [vertices], isClosed = True, color = (255,0,0), thickness = 5)
plt.imshow(img6)
```

For Triangle, pls mention only 3 vertices
