#### HSL - Hue, Saturation and Lightness, also called as HSL
#### HSV - Hue, Saturation and Value
#### Hue is 0 degrees to 360 degress, color changes from VIBGYOR
#### Saturation will control color presence, 0% turns to gray scale and 100% to colorful (True Color) image
#### Lightness - between 0 and 1 - 0 is Dark and 1 is full bright


HSV, H and S - same 
V - Value - Between 0 and 1 - 0 is dark and 1 is True Color 


```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./hummingbird.jpg')
plt.imshow(img)
```

Image is in BGR colorspace. Convert it into RGB

```python
pic = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(pic)
```

Now convert the image to HSV

```python
pic2 = cv2.cvtColor(pic, cv2.COLOR_RGB2HSV)
plt.imshow(pic2)
```

```python
pic3 = cv2.cvtColor(pic, cv2.COLOR_RGB2HLS)
plt.imshow(pic3)
```

Converting BGR directly into HSV and HSL(HLS)

```python
pic4 = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(pic4)
```

```python
pic5 = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
plt.imshow(pic5)
```
